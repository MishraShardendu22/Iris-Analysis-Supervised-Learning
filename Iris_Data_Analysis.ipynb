```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
df.drop_duplicates()
df["species"].unique()
```

```python
xDf = df.drop(columns=['species'])
yDf = df['species']
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cols = xDf.columns
idx = xDf.index

xDf = pd.DataFrame(
    scaler.fit_transform(xDf),
    columns=cols,
    index=idx
)

xDf.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(xDf, yDf, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn = KNeighborsClassifier(n_neighbors=13)
model_knn.fit(X_train, y_train)
model_knn.score(X_test, y_test)
```

```python
from sklearn.svm import SVC

model_svc = SVC(gamma='auto')
model_svc.fit(X_train, y_train)
model_svc.score(X_test, y_test)
```

```python
# Using grid-search CV

from sklearn.model_selection import GridSearchCV

classifier = GridSearchCV((model_svc),{
    'C' : [1,10,20,30],
    'kernel' :['rbf','linear'],
},cv = 5,return_train_score = False)

classifier.fit(xDf,yDf)
df_classifier = pd.DataFrame(classifier.cv_results_)
df_classifier
```

```python
df_classifier[['param_C', 'param_kernel', 'mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV

classifier_r = RandomizedSearchCV(
    model_svc,
    param_distributions={
        'C': [1, 10, 20, 30],
        'kernel': ['rbf', 'linear']
    },
    n_iter=4,
    cv=5,
    return_train_score=False,
    random_state=42
)

classifier_r.fit(xDf, yDf)

df_classifier_r = pd.DataFrame(classifier_r.cv_results_)
df_classifier_r
```

```python
df_classifier_r[['param_C', 'param_kernel', 'mean_test_score', 'rank_test_score']]
```
